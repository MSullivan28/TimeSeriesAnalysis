### Model Fitting 

```r
### AICC
arma.AICC <- function(fit) {
  n.par <- length(fit$coef)
  n <- fit$nobs
  aicc <- -2*fit$loglik+2*(n.par+1)*n/(n-n.par-2)
  return(aicc)
}
```

### Model Diagnostics

```r
x.t <- usrent.rsd.t 

fit.21B <- arima(x.t,order=c(2,0,1),include.mean=FALSE,method="CSS-ML")
fit.12B <- arima(x.t,order=c(1,0,2),include.mean=FALSE,method="CSS-ML")

fit.bst <- fit.21B
w.t <- fit.bst$resid 
w.mean <- mean(w.t)                                
w.sd <- sd(w.t)                                    
w.mean
w.sd
m <- 10                                                           # Ljung-Box portmanteau test's "m"
n.par <- length(fit.bst$coef)                                     # number of parameters in fitted ARMA(p,q), p+q+1
LB.out <- Box.test(w.t,lag=m,type="Ljung-Box")                    # Ljung-Box portmanteau test
LB.out                                                            # Box.test uses wrong df if applied to residuals
LB.pval <- pchisq(q=LB.out$statistic,df=m-n.par,lower.tail=FALSE) # correct p-value
LB.pval

x.us <- fit.bst
```
